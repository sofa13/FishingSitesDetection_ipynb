# Hosting providers asn list

```python
updated_hosting_service_fout = "./domainclassification/features/hostingService/hosting_providers_2"

# hosting service info
with open(updated_hosting_service_fout, 'r', encoding="utf-8") as fin:
    hosting_providers = eval(fin.read())
    
len(hosting_providers)
```

```python
from FishingSitesDetection.PhishEnd.Common import URLResolver
from pprint import pprint

#hosting_providers = list(hosting_providers)[:2]
ur = URLResolver()
domain_to_asn = dict()
for host in hosting_providers:
    ip = ur.getIP(host)
    if ip:
        ipinfo = ur.getIPInfo(ip)
        
        if ipinfo and 'as' in ipinfo:
            try:
                asn = ipinfo['as'].split(' ')[0].split('AS')[1]
            except Exception as e:
                print(e)
                print(ipinfo)
            else:
                domain_to_asn[host] = asn

pprint(domain_to_asn)
```

```python
len(domain_to_asn)
```

```python
fnout = "./domainclassification/features/hostingService/hosting_providers_to_asn"

# hosting service info
with open(fnout, 'w', encoding="utf-8") as fout:
    pprint(domain_to_asn, fout)
```
