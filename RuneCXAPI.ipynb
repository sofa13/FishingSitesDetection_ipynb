```python
from FishingSitesDetection.Sophie.API.eCXAPI import eCXAPI as ecxa

import requests
import json
import sys
import datetime
from pprint import pprint
import os
```

# Get phish data

```python
ecx = ecxa()

years = [2018]
months = [5, 6, 7, 8, 9, 10, 11, 12]
days = {5: 30, 6: 31, 7: 30, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}

timestamp = None
file = None
folder = "./domainclassification/groundtruth_data/ecrimex/phish2/"
for yr in years:
    for mn in months:
        dys = days[mn]
        for dy in range(1,dys):
            
            dt = datetime.datetime(yr, mn, dy, tzinfo=datetime.timezone.utc)
            epoch = datetime.datetime(1970, 1, 1, tzinfo=datetime.timezone.utc)
            next_timestamp = (dt - epoch) // datetime.timedelta(seconds=1)
            
            # start may 1st 2018 end jan 1st 2019
            if timestamp is not None:
                endpoint = '/phish?dd_date_start='+str(timestamp)+'&dd_date_end='+str(next_timestamp)+'&confidence_level=100&order_by=date_discovered&order=asc'

                entities = []
                while endpoint and timestamp is not None:
                    #print(endpoint)
                    res = ecx.call(endpoint)

                    entities += res['_embedded']['phish']

                    endpoint = res['_links'].get('next', None)
                    if endpoint:
                        endpoint = endpoint.get('href', None)
                        
                        # there may still be more...
                    elif len(entities)%10000==0:
                        timestamp = entities[-1]['date_discovered']
                        endpoint = '/phish?dd_date_start='+str(timestamp)+'&dd_date_end='+str(next_timestamp)+'&confidence_level=100&order_by=date_discovered&order=asc'

                print(len(entities))

                if file:
                    print(file)
                    file = os.path.join(folder, file)
                    with open(file, 'w', encoding="utf-8") as fout:
                        pprint(entities, fout)
            
            timestamp = next_timestamp
            file = datetime.datetime.utcfromtimestamp(timestamp).strftime('%Y-%m-%d')
```
